```python
import sys
sys.path.append('../')
from plot_tools.fig_preamble import *

from matplotlib.ticker import FormatStrFormatter
```

```python
ver=0
datapath = './DATA_v'+str(ver)+'_/' #v'+str(ver)+'/DATA_v'+str(ver)+'/'
saveto = './FIGS_v'+str(ver)+'/' #v'+str(ver)+'/FIGS_v'+str(ver)+'/'
savefig = './FIGS_v'+str(ver)+'/poster_fig4_'
```

```python
nrows = 3
ncols = 6
test_case_list = [27, 19, 19]
variation_list = [5, 5, 8]
ylim_cases = [-0.1, 0, 0]
noise_lvl = [10, 10 , 25]
# Dump unwanted axes
dumpfig = plt.figure(1)
dumpax = dumpfig.add_subplot(111)


# Fig Template
gs = gridspec.GridSpec(nrows, ncols,
                       left=0.2, right=0.97, 
                       top=0.93, bottom=0.1, 
                       wspace=10, hspace=0.5)
fig1 = plt.figure(figsize=(cm2inch(6.), cm2inch(6.)), dpi=400)


for idx_c in xrange(3):
    vars()[str(idx_c)+'_ax1'] = fig1.add_subplot(gs[idx_c,0:4])
    vars()[str(idx_c)+'_ax2'] = fig1.add_subplot(gs[idx_c,4:])
    
    if idx_c != 2:
        algorithms = ALGOKEYS[0:3]
    elif idx_c ==2:
        algorithms = ALGOKEYS[0:2]


    vars()[str(idx_c)+'_ax1'], dumpax = psp(vars()[str(idx_c)+'_ax1'], dumpax, algorithms, 
                 test_case_list[idx_c], variation_list[idx_c], 
                 datapath, GPRP_load='No',
                 fstep=50, sstep=50, lowloss=20, nt_label=15,
                 ylim_amps = [-14, 2], yscale='linear')

    dumpax, vars()[str(idx_c)+'_ax2'] = pnm(dumpax, vars()[str(idx_c)+'_ax2'],  algorithms, 
                    test_case_list[idx_c], variation_list[idx_c], 
                    datapath, GPRP_load='No',
                    fstep=50, sstep=50, lowloss=20, 
                    ylim = [-4.5, 3.75], yscale='log')

    for ax in [vars()[str(idx_c)+'_ax1'], vars()[str(idx_c)+'_ax2']]:
        ax.set(xlabel='', ylabel='')
        ax.tick_params(direction='in', axis='both')
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        #ax.yaxis.set_major_formatter(FormatStrFormatter('%0.f'))

    vars()[str(idx_c)+'_ax2'].yaxis.set_major_formatter(FormatStrFormatter('%0.f'))
    vars()[str(idx_c)+'_ax2'].locator_params(nbins=3, axis='y')
    vars()[str(idx_c)+'_ax2'].set_ylim([ylim_cases[idx_c],2])
    vars()[str(idx_c)+'_ax2'].margins(0.1)
    
    #plt.text(0.7, 1.2, r'%s'%(noise_lvl[idx_c]), fontsize=8, horizontalalignment='right',
    #         verticalalignment='center', transform=vars()[str(idx_c)+'_ax1'].transAxes)
    ##
    
    if idx_c != 2:
        vars()[str(idx_c)+'_ax1'].set_xticklabels([])#xaxis.set_visible(False)
        vars()[str(idx_c)+'_ax2'].set_xticklabels([])#xaxis.set_visible(False)
        vars()[str(idx_c)+'_ax1'].spines['bottom'].set_position('zero')
        vars()[str(idx_c)+'_ax1'].tick_params(direction='inout', axis='x') 
```

```python
fig1.savefig(savefig+'test.svg', format='svg', dpi=400)
plt.show()
plt.close()
```
